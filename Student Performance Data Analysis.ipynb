```python
# All countplots

# gender barplot
sns.countplot(x = df['gender'], palette ="Blues_d")
plt.figure(figsize=(14,9))
plt.show()

# race / ethnicity bar plot
sns.countplot(x = df['race/ethnicity'], palette='husl')
plt.figure(figsize=(14,9))
plt.show()

# education level 

sns.countplot(x = df['parental level of education'],palette='magma')
plt.figure(figsize=(19,19))
plt.show()

# test preparation course

sns.countplot(x = df['test preparation course'],palette='crest')
plt.figure(figsize=(14,9))
plt.show()

# lunch count 
sns.countplot(x = df['lunch'])
plt.figure(figsize=(14,9))
plt.show()
```

```python
# pie chart for education
sns.set_palette ('magma')
keys=education.index
sizes = education.values
plt.pie(sizes , labels=keys,autopct='%1.1f%%')
plt.show()
```

```python
# count for race and ethnicty 
sns.set_palette('bright')
plt.figure(figsize=(12,9))
sns.countplot(x= df['race/ethnicity'])
plt.title('Groups and values')
plt.show()
```

```python
# bar plots

sns.barplot(x = df['race/ethnicity'].value_counts().index, y=df['race/ethnicity'].value_counts())
plt.figure(14,7)
plt.xlabel('Gender')
plt.show()
```

```python
# bar plot scores
plt.figure(figsize=(16, 6))
sns.barplot(x=df['math score'].value_counts().index, y=df['math score'].value_counts())
plt.title('acquired math scores')
plt.xticks(rotation=90)
plt.show()
```

```python
# count plot chart
sns.set_palette('dark')
plt.figure(figsize=(12,9))
sns.countplot(x='parental level of education', hue='gender', data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
# calculate correlation
correlation = df.corr()
sns.heatmap(vmin =-1.0 , vmax =1.0 , data = correlation ,cmap='viridis' )
plt.title('Correlation')
plt.show()
```

```python
# scatter plot
sns.scatterplot(data=df, x ='math score' , y ='reading score', palette = 'bright')
plt.title('Scatter Plot')
plt.show()
```

```python
# scatter plot
sns.scatterplot(data=df, x ='reading score' , y ='writing score', hue = 'gender', palette = 'coolwarm')
plt.title('Scatter Plot')
plt.show()
```

```python
# bar plot
sns.bar()
plt.title('Bar Chart')
plt.show()
```

```python
# pairplot by gender
sns.pairplot(df, hue = 'gender')
plt.show()
```
